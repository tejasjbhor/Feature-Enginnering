```python
import pandas as pd
import numpy as np

# take only categorical variables as we requrired only few

dataset = pd.read_csv('H:/Machine Learning/mercedes-benz-greener-manufacturing/mercedes.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
dataset.head()
```

```python
# lets count how many labels each variable contains

for col in dataset.columns:
    total=len(dataset[col].unique())
    print(col,'has',total,'labels')
```

```python
# if we perform one_hot_encoding on all the above features , the how many columns will get created

#O/P is 117 columns created total from 6 columns

pd.get_dummies(dataset, drop_first=True).shape
```

<b>Too many columns increses that leads to curse of dimentionality, so  insted of getting too many columns 
WE WILL TAKE 10 MOST FREQUENT CATEGORIES FROM EACH FEATURE and perform one hot encoding<b/>

count of 10 may change depends upon the categories

```python
#foe an ex. lets take col X2

dataset.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# lets make list with most frequent categories of the variable

top_10 = [x for x in dataset.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# create 10 binary variables

for label in top_10:
    dataset[label] = np.where(dataset['X2']==label, 1, 0) 
    
    # if X2 is equal to Label in top_10 then replace with 1 or keep it as 0

dataset[['X2']+top_10].head(2)
```

```python
# now we will impliment this in top 10 most frequent features and convert it to dummy variables

def one_hot_encoding_X(df, variable, top_x_labels):
    # function to create dummy variable for the most frequent labels
    
    for label in top_x_labels:
        dataset[variable+'_'+label] = np.where(dataset[variable]==label, 1 ,0)        

# again read dataframe and check with function

dataset = pd.read_csv('H:/Machine Learning/mercedes-benz-greener-manufacturing/mercedes.csv', 
                      usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

one_hot_encoding_X(dataset,'X2',top_10)

dataset.head()
```

```python
# similerly for X1 we perform as follows

top_10 = [x for x in dataset.X1.value_counts().sort_values(ascending=False).head(10).index]
#top_10_x1

one_hot_encoding_X(dataset,'X1',top_10)

dataset.head()
```
